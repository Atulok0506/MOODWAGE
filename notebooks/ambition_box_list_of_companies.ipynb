```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers ={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
#web_scrap=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
rating = []
overview = []
reviews = []
company_names = []
logo_url = []
logo_urls=[]
```

```python
from IPython.display import Image, display
```

```python
# Loop through the pages
for i in range(1, 300):
    web_scrap = requests.get(f'https://www.ambitionbox.com/list-of-companies?page={i}', headers=headers).text
    soup = BeautifulSoup(web_scrap, 'html.parser')
    
    # Find and extract company information
    comp = soup.find_all('div', class_='companyCardWrapper__metaInformation')
    for c in comp:
        rating.append(c.find('span', class_='companyCardWrapper__companyRatingValue').text.strip())
        overview.append(c.find('span', class_='companyCardWrapper__interLinking').text.strip())
    # Find and extract review information
    tertiary = soup.find_all('div', class_='companyCardWrapper__tertiaryInformation')
    for t in tertiary:
        review_span = t.find('span', class_='companyCardWrapper__ActionCount')
        if review_span:
            reviews.append(review_span.text.strip())
    names = soup.find_all('h2', {'class': 'companyCardWrapper__companyName'})
    for name in names:
        company_name = name.get_text(strip=True)
        company_names.append(company_name)
    
    # Find and extract logo URLs
    logo_elements = soup.find_all('div', class_='companyCardWrapper__companyLogo')
    for logo_element in logo_elements:
        logo_url = logo_element.find('img')['src']
        logo_urls.append(logo_url)
        
        display(Image(url=logo_url))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import os
import time

# Directory to save logo images
logo_directory = 'logo_images_3'

# Ensure the directory exists, or create it if it doesn't
if not os.path.exists(logo_directory):
    os.makedirs(logo_directory)

# Initialize the Selenium WebDriver (make sure to download the appropriate driver for your browser)
driver = webdriver.Chrome()  # You can use other web drivers like Firefox, etc.

# Loop through the logo URLs and save the images using Selenium
for idx, (logo_url, company_name) in enumerate(zip(logo_urls, company_names)):
    try:
        # Open the URL in the browser
        driver.get(logo_url)
        
        # Create a unique filename for each image using the company name
        filename = f'{company_name}_{int(time.time())}_{idx}.jpg'
        filepath = os.path.join(logo_directory, filename)

        # Save the image from the browser
        with open(filepath, 'wb') as file:
            file.write(driver.find_element(By.TAG_NAME, 'img').screenshot_as_png)
        
        print(f"Image {idx + 1} for {company_name} saved as {filename}")
    except Exception as e:
        print(f"Failed to download image for index {idx} with URL {logo_url}: {str(e)}")

# Close the browser
driver.quit()
```

```python
extracted_info_list = []

for overview_str in overview: 
    info_parts = overview_str.split('|')
    info_parts = [part.strip() for part in info_parts]
    
    # Ensure that each part of the info_parts list corresponds to the correct column
#     industry = info_parts[0]
#     employee_count = None if len(info_parts) <= 1 else info_parts[1]
    
    # Handle cases where Ownership and Location values are mixed up with Age
    employee_count = None
    ownership = None
    location = None
    age = None
    for part in info_parts[0:]:
        if "Employees" in part: # or "None" in part:
            employee_count = part
        elif "Public" in part or "Forbes" in part or "Conglomerate" in part:
            ownership = part
        elif "years old" in part:
            age = part
        elif "more" in part:
            location = part
        else:
            industry = part

    extracted_info_list.append({
        'Industry': industry,
        'Employee Count': employee_count,
        'Ownership': ownership,
        'Age': age,
        'Location': location
    })

# Now you can create a DataFrame from the extracted information
import pandas as pd

df = pd.DataFrame(extracted_info_list)
print(df.sample(10))
```
